```python
from sklearn.metrics.pairwise import euclidean_distances
import pandas as pd

# "Coldplay - Yellow" spotify_id
coldplay_spotify_id = '3AJwUDP919kvQ9QcozQPxg'

# Check if the given spotify_id exists in x_scaled_df_spotify
matching_song = x_scaled_df_spotify[x_scaled_df_spotify['spotify_id'] == coldplay_spotify_id]

if not matching_song.empty:
    # Extract the audio feature data for "Coldplay - Yellow"
    shape_of_you_features = matching_song.drop(['Unnamed: 0', 'spotify_id', 'cluster'], axis=1).values.reshape(1, -1)
    
    # x_scaled_df_spotify
    
    # Calculate similarity with other songs (using Euclidean distance)
    similarity_matrix = euclidean_distances(shape_of_you_features, x_scaled_df_spotify_relevant_cluster.drop(['Unnamed: 0', 'cluster', 'spotify_id'], axis=1).values)

    # Find top 5 songs with highest similarity
    top_5_similar_songs = similarity_matrix[0].argsort()[1:6]  # Exclude the song itself (index 0)

    # Print the recommended songs
    # recommended_songs = x_scaled_df_spotify.iloc[top_5_similar_songs]
    recommended_songs = x_scaled_df_spotify_relevant_cluster.iloc[top_5_similar_songs]
    
    # Retrieve Track and Album information for the recommended spotify_ids from df_final_aggregated
    recommended_songs_info = df_final_aggregated[df_final_aggregated['spotify_id'].isin(recommended_songs['spotify_id'])]

    print("Recommended songs:")
    print(recommended_songs_info[['Album','Title', 'spotify_id']])  # Print Track, Album, spotify_id of recommended songs
else:
    print("The given spotify_id is not in the data.")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Perform t-SNE dimensionality reduction on all songs in x_scaled_df_spotify
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(x_scaled_df_spotify_relevant_cluster.drop(['Unnamed: 0', 'spotify_id', 'cluster'], axis=1))

# Extract indices of the recommended songs
recommended_tsne = X_tsne[top_5_similar_songs]

# Find the index of "Coldplay - Yellow"
coldplay_index = x_scaled_df_spotify[x_scaled_df_spotify['spotify_id'] == '3AJwUDP919kvQ9QcozQPxg'].index[0]
coldplay_tsne = X_tsne[coldplay_index]  # t-SNE coordinates for "Coldplay - Yellow"

# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c='blue', label='All Songs within the cluster', alpha=0.5)  # All songs
plt.scatter(recommended_tsne[:, 0], recommended_tsne[:, 1], c='red', label='Recommended Songs', marker='X')  # Recommended songs
plt.scatter(coldplay_tsne[0], coldplay_tsne[1], c='green', label='"Coldplay - Yellow"', marker='D', s=100)  # Highlight "Coldplay - Yellow"
plt.title('t-SNE Visualization of Recommended Songs')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()
plt.show()
```
