```python
import pandas as pd

# Ensure all rows and columns are displayed
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
# pd.set_option('display.max_colwidth', None)
```

```python
df_song_attrs_lyrics = pd.read_csv("songs_with_attributes_and_lyrics[1].csv")
```

```python
df_spotify_youtube_comments = pd.read_csv("spotify_youtube_comment_88247.csv")
```

```python
df_spotify_youtube = pd.read_csv("Spotify_Youtube.csv")
```

```python
df_spotify_youtube.head()
```

```python
for col in df_spotify_youtube.columns:
    print(col)
```

```python
df_spotify_youtube['spotify_id'] = df_spotify_youtube.Uri.apply( lambda x: x.split(":")[-1] )
```

```python
df_spotify_youtube[['spotify_id', 'Uri']].head()
```

```python
reqd_columns = [
    "Artist", "Url_spotify", "Track", "Album", "Album_type", "Uri", 
    "Danceability", "Energy", "Key", "Loudness", "Speechiness", 
    "Acousticness", "Instrumentalness", "Liveness", "Valence", 
    "Tempo", "Duration_ms", "Url_youtube", "Title","spotify_id"
]
```

```python
df_spotify_youtube = df_spotify_youtube[reqd_columns]
```

```python
df_song_attrs_lyrics.head()
```

```python
merged_df = df_spotify_youtube.merge(df_song_attrs_lyrics[['id', 'lyrics']], left_on="spotify_id", right_on="id", how="left")
```

```python
len(merged_df), len(df_spotify_youtube)
```

```python
merged_df.dropna(subset=['spotify_id', 'Url_youtube', 'id'], inplace=True)
len(merged_df)
```

```python
merged_df.head()
```

```python
df_spotify_youtube_comments.head()
```

```python
merged_df['youtube_video_id'] = merged_df.Url_youtube.apply( lambda x : x.split("?v=")[-1] )
```

```python
len(merged_df)
```

```python
merged_df.to_csv("merged_df.csv")
```

```python
spotify_youtube_comments_combined = merged_df.merge(
    df_spotify_youtube_comments, left_on = ["spotify_id", "youtube_video_id"], right_on = ['spotify_id', 'video_id'], how="left"
)

```
